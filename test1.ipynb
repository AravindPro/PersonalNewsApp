```python
from sklearn.cluster import DBSCAN
import numpy as np
from sklearn.datasets import make_blobs
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import pairwise_distances


def custom_distance(x, y):
    return np.dot(x, y)


# Calculate pairwise distances using custom distance metric

MODEL = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2", cache_folder="./cache")
```

```python
from trafilatura import fetch_url, extract

def getArticle(url):
	downloaded = fetch_url(url)
	res = extract(downloaded)
	# print(res)
	return res
```

```python
# Article similarity
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cdist
def cosine(u,v):
	u = u.reshape(1, -1)
	v = v.reshape(1, -1)
	return cosine_similarity(u, v)
# Assuming you have a matrix 'data' of shape (n, d)
# data = np.array([[x1, y1, z1], [x2, y2, z2], ...])  # Example data
cosine_similarities = cdist(a2, a2, metric=cosine)
np.set_printoptions(precision=2, suppress=True)
print(cosine_similarities)
```

```python

# # X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
# # print(type(X))
# # Instantiate DBSCAN clustering object
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(pairwise_dist)

labels = dbscan.labels_

# # Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
print(labels, n_clusters_, n_noise_)
```
